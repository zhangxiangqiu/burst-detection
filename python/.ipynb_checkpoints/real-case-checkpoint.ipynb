```python
# 压力数据，数据预处理

def time2str(time):
    return time.strftime("%Y-%m-%d %H:%M:%S")

def str2time(s):
    return datetime.strptime(s, '%Y/%m/%d %H:%M')


data = pd.read_csv(r'data/flow/XK0601.csv', header=0) 
data = data.drop(['SITENAME', 'NAME'],axis=1) #axis =1列 axis=0行
data['datetime'] = data['DATE'] + ' ' + data['TIME']
data['datetime'] = data['datetime'].apply( str2time ).apply(time2str)
tmp = data['datetime'].str.split(' ', expand=True)
data = pd.concat([tmp, data], axis=1)
data['val'] = data['CHECKVALUE']
data = data.drop(['DATE', 'TIME', 'CHECKVALUE'], axis=1)
data.columns = ['date', 'time', 'datetime', 'val']
data.drop_duplicates(subset='datetime', keep='first', inplace=True) 
#duplicate是删除重复的数据  first 将第一次出现重复值标记为true，inplace删除之后替代
data = data.reset_index(drop=True) #修改排序
# data.to_csv(r'data/flow/XK0601_processed.csv', header=None, index=None)
data
```

```python
def get_time_list(start, length):
    time_delta = timedelta(minutes=5)
    time = datetime.strptime(start, "%Y-%m-%d %H:%M:%S") - time_delta #strptime 将时间字符串转时间
    _time_list = []
    while True:
        time += time_delta
        #         time_str = time.strftime("%Y-%m-%d %H:%M:%S.%f")[:-3]
        time_str = time.strftime("%H:%M:%S") #strftime 将时间转成字符串
        _time_list.append(time_str)
        if len(_time_list) == length:
            break
    return _time_list
time_list = get_time_list("2005-01-01 00:00:00", 288)

for time_index in time_list: #在2sigma之外的数据用mean代替 处理真实数据的时候剔除一些异常值很大的值
    mean = data[data['time']==time_index]['val'].mean()
    std = data[data['time']==time_index]['val'].std()
    data.loc[data[(data['time']==time_index) & (data['val'] > mean+2*std)].index, 'val'] = mean
    data.loc[data[(data['time']==time_index) & (data['val'] < mean-2*std)].index, 'val'] = mean
data.to_csv(r'data/flow/XK0601_fuben.csv', header=None, index=None)
#删掉了2sigma之外的数据
```

```python
#划分好窗口，为随机森林算法模型准备数据（人为设定好某个时间，将某个时间的数据库提取，下一部分显示时间）
window_size = 6
slide_size = 0

d = {}

for i in tqdm(range(np.shape(data)[0])):  #tqdm是进度条 np.shape(data)[0]是行数 i是第几行
    time = data.iloc[i, 1] #第一列 data.iloc[i,1] 是时间00:00:00 定位到具体时间那一列
    for j in range(i-slide_size, i+slide_size+1):  # j是窗口的末尾时刻 由于漂移量，j时刻的数据多个
        start = j-window_size+1
        end = j    #j时刻是窗口末尾时刻
        if start < 0 or end >= np.shape(data)[0]: 
            continue
        x = data.iloc[start:end+1, 3].values   # 定位到时间步长，哪一个时刻到哪个时刻的数值，3是0 123 第四列
        sample = [x.min(), x.ptp(), x.std(), np.max(np.abs(np.diff(x)))]  #删除x.std(),  将sum改成max
        if time not in d: #d是空的，时间不是空的，则时间=空的，
            d[time] = []
        d[time].append(sample) # 在time后面加入sample 同一时刻的sample放在一起

#变化值的绝对值之和
```

```python
#利用孤立森林进行训练同一时刻的数据：1 建立初始的数据库，设定异常比例，将前面最异常的踢掉，再不断进入样本
#当样本库达到一定数量的时候，更新数据库，将检测到的正常样本，把时间最老的删掉；异常的直接删掉（初始60 最大80，达到最大以后就进行数据库更新）
#效果和异常比例有关系，异常比例决定初始数据库筛掉的值，比例设置大小会影响初始数据库的样本特征，样本比例太大，容易忽略正常的波动
#IF算法步骤是准备数据+fit模型+预测+结果
init_sample_num = 60  #初始数据库
max_sample_num = 80   #最大数据库
alpha = 0.1 #异常比例
time = '02:10:00'

X = pd.DataFrame(d[time], columns=['max', 'min', 'ptp', 'std', 'abs_diff'])
model = IsolationForest(n_estimators=300, contamination=alpha, random_state=42)
#contamination 为 the proportion of outliers 。在拟合时用于定义样本分数的阈值 
X_init = X.iloc[0:init_sample_num, :].values #定位到某个值，从0定位到初始数据库的数量 
model.fit(X_init)#拟合
scores = model.predict(X_init) #预测初始数据库的分值
X_train = X_init[scores == 1] #看看初始数据库中，分数是否为1，为1的话作为训练集，将其他的排除
res = np.zeros(len(X)) #将x的长度用0返回
# 上面是将初始数据库按照异常比例alpha排除异常的(contamination=alpha)，将正常的筛选出来作为训练集，300颗树
# 下面是没有异常比例
model = IsolationForest(n_estimators=300, contamination=0, random_state=42)

for i in tqdm(range(init_sample_num, len(X))): #i在最初的数据库到x的长度 
    model.fit(X_train) #拟合训练集 
    cur_sample = X.iloc[i, :].values.reshape(1, -1) #reshape(1,-1) 转换成一行
    pred = model.predict(cur_sample) #预测
    res[i] = pred[0] 
    if pred[0] == 1: #如果预测的是正常的
        if len(X_train) < max_sample_num:
            X_train = np.append(X_train, cur_sample, axis=0)
        else:
            X_train = np.append(X_train[1:], cur_sample, axis=0) #训练集就是从第一个到最后  axis=0是纵向
res
#初始库异常比例alpha
```

```python
init_sample_num = 60  #初始数据库
max_sample_num = 80   #最大数据库
alpha = 0.2 #异常比例
time = '02:20:00'
pre_seen = 4

def get_time_list(start, length):
    time_delta = timedelta(minutes=5)
    time = datetime.strptime(start, "%Y-%m-%d %H:%M:%S") - time_delta
    _time_list = []
    while True:
        time += time_delta
        #         time_str = time.strftime("%Y-%m-%d %H:%M:%S.%f")[:-3]
        time_str = time.strftime("%H:%M:%S")
        _time_list.append(time_str)
        if len(_time_list) == length:
            break
    return _time_list
time_list = get_time_list("2005-01-01 01:00:00", 25)

X_train = {}
X = {}
for time in time_list:
    X[time] = pd.DataFrame(d[time], columns=['max', 'min', 'ptp', 'std', 'abs_diff'])
    model = IsolationForest(n_estimators=100, max_samples=0.6, max_features=0.6, contamination=alpha, random_state=42)
    X_init = X[time].iloc[0:init_sample_num, :].values
    model.fit(X_init)
    scores = model.predict(X_init)
    X_train[time] = X_init[scores == 1]

res = {}
for time in time_list:
    res[time] = np.zeros(len(X[time]))
        
model = IsolationForest(n_estimators=300, max_samples=0.6, max_features=0.6, random_state=42)
for j in tqdm(range(len(time_list))):
    for i in range(init_sample_num, len(X[time_list[j]])):
        isAbnormal = True
        for k in range(pre_seen):
            model.fit(X_train[time_list[j-k]]) #拟合训练集 
            cur_sample = X[time_list[j-k]].iloc[i, :].values.reshape(1, -1) #reshape(1,-1) 转换成一行
            pred = model.predict(cur_sample) #预测
            if pred[0] == 1:
                isAbnormal = False
                break
        res[time_list[j]][i] = -1 if isAbnormal else 0
        if not isAbnormal: #如果预测的是正常的
            if len(X_train[time_list[j]]) < max_sample_num:
                X_train[time_list[j]] = np.append(X_train[time_list[j]], cur_sample, axis=0)
            else:
                X_train[time_list[j]] = np.append(X_train[time_list[j]][1:], cur_sample, axis=0) #训练集就是从第一个到最后  axis=0是纵向
res
#往下是自己写的
# if len(X_train) > max_sample_num and pre[0]:
    
#连续几个才是异常的
```
