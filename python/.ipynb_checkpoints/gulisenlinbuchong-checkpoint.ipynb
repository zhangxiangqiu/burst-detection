```python
#生成一个96*10的数据库，96是一天多少个步长，10是压力计个数，写入centroidPressure文件中
def PearsonSimilarity(x, y):
    '''
    皮尔森相似度
    '''
    return pearsonr(x,y)[0]

def EuclideanDistance(x, y): #距离
    '''
    欧几里得相似度
    '''
    return np.sqrt(sum(pow(a-b,2) for a,b in zip(x,y)))

def ManhattanDistance(x, y):
    '''
    曼哈顿相似度
    '''
    return sum(abs(a-b) for a,b in zip(x,y))
    
def CosineSimilarity(x, y):
    '''
    余弦相似度
    '''
    tmp=np.sum(x*y)
    non=np.linalg.norm(x)*np.linalg.norm(y)
    return np.round(tmp/float(non),9)


def CustomSimilarity(burst, centroid): #自定义相似性(形状上描述)
    base = burst/centroid #实际数据/质心数据（burst是含有爆管的未知实时数据，不是爆管数据）
    res = 0
#先进行窗口中步长之间计算，再求绝对值
    #res_final = 0
    #for i in range(1, len(base)):
        #res += (base[i]-base[0])/ i     #绝对值！！与第一个时刻做差，然后除以与第一个时刻的差值 (b-a)/1+(c-a)/2+(d-a)/3
        #res_final = abs(res)
    #return res_final  
#计算窗口中步长之间的计算，就得是绝对值
    for i in range(1, len(base)): 
        res += abs(base[i]-base[0]) / i 
    return res
```

```python
#第一步！！重新准备一份burstpressure数据集，改成burstpressureProcess，
#能够进行处理。获得新的customProcess和EuclideanProcess文件

#原有的centroidpressure。后面加个Process是得到每个时刻（欧氏距离）和（custom）的最大最小平均值
#为了让每个时刻差不多。越异常距离越大，越正常距离越小
centroidPressureFile = 'resource/centroidPressure.csv'
centroidPressure = pd.read_csv(centroidPressureFile, header=None) 
print(centroidPressure.shape) 
#重新生成burstpressure，将burstpressure文件改成burstpPressureProcess，用该数据与centroidpressure进行计算
burstPressureProcessFile = 'resource/burstPressureProcess.csv' 
burstPressureProcess = pd.read_csv('resource/burst_pressure_Process.csv', header=None)
print(burstPressureProcess.shape) 
#将质心数据与数据表进行
centroidPressure = pd.DataFrame(pd.np.tile(centroidPressure, (burstPressureProcess.shape[0] // centroidPressure.shape[0], 1)))
#打印重复之后的质心压力表的行列数
print(centroidPressure.shape)
windowSize=3 #第一二行是0，因为时间序列数据是3个步长
#生成的欧式距离、自定义距离表格和上面合并之后的burstpressure表格行列数相同

euclideanProcess = pd.DataFrame(np.zeros(burstPressureProcess.shape))#
customProcess = pd.DataFrame(np.zeros(burstPressureProcess.shape))#生成的custom表格和burstpressure表格行列数相同
#tqdm进度条 shape[1]是列数
for col in tqdm(range(burstPressureProcess.shape[1])):#  列数 
#第一行在代码里是0，第二行是1，窗口是3，从第三行到burstpressure总行数 以窗口为单位计算！！！
    for row in range(windowSize-1, burstPressureProcess.shape[0]): 
        centroid = centroidPressure.iloc[row-windowSize+1:row+1, col].values #质心数据 将原始的以时刻为单位进行划分
        burstProcess = burstPressureProcess.iloc[row-windowSize+1:row+1, col].values #爆管数据，将原始的以时刻为单位进行划分
        #row-windowSize+1:row+1 左边包括右边不包括，row是行(时刻) 如果是第五个时刻raw=5 windows=4 [2,6) 2,3,4,5
        euclideanProcess.iloc[row, col] = EuclideanDistance(centroid, burstProcess)#将上面划分好的窗口计算欧式距离 
        customProcess.iloc[row, col] = CustomSimilarity(centroid, burstProcess)#将上面划分好的窗口计算自定义距离
print(euclideanProcess.head())
print(customProcess.head()) 
#生成处理好的距离
```

```python
#生成结果文件
trainBurstLabel = pd.read_csv('resource/train_burst_label.csv', header=None) #训练集的爆管标签
burstLabel = pd.read_csv(burstLabelFile, header=None) #测试集和验证集的爆管标签，读取文件
burstLabel = pd.concat([trainBurstLabel,burstLabel]).iloc[:, 0].values #concat是合并，trainBurstLabel训练集标签，burstlabel测试集和验证集标签
print(burstLabel.shape) #打印出合并后burstlabel的shape #iloc[:, 0]是所有行第一列的数据
y_true = burstLabel[windowSize-1:] #比如窗口是3，则标签应该是从第三个开始 
y_pred = model.predict(X) #预测数据
y_pred = np.array(list(map(lambda x: 1 if x == -1 else 0, y_pred)))#x与-1相当，则等于1，否则等于0
print(y_pred.shape, y_true.shape) #if等于-1
#splitIndex1 = len(trainBurstPressure)
#splitIndex2 = len(trainBurstPressure) + math.ceil(len(testBurstPressure)*0.5)

toPrintDF = X
toPrintDF['y_true'] = y_true 
toPrintDF['y_pred'] = y_pred
def getResult(y_true, y_pred):
    res = []
    for i in tqdm(range(len(y_true))):
        t = y_true[i]
        p = y_pred[i]
        if t == 1 and p == 1:  #1是爆管，0是正常  是爆管检测出爆管
            res.append('TP') # true positive
        elif t == 0 and p == 0:#正常，检测出正常
            res.append('TN')
        elif t == 0 and p == 1: #正常，检测出爆管
            res.append('FP')
        else: # t == 1 and p == 0 #爆管，检测出正常
            res.append('FN')
    return res

toPrintDF['result'] = getResult(toPrintDF['y_true'], toPrintDF['y_pred'])
resultTrain = toPrintDF.iloc[:splitIndex1, -1].values #训练集的结果
resultVal = toPrintDF.iloc[splitIndex1:splitIndex2, -1].values #验证集的结果
resultTest = toPrintDF.iloc[splitIndex2:, -1].values #测试集的结果

def countResult(label, data):
    counter = Counter(data)
    print(counter)
    tpr = counter['TP'] / (counter['TP'] + counter['FN'])
    fpr = counter['FP'] / (counter['FP'] + counter['TN'])
    print(label, ': TPR = TP / (TP + FN)：', '{:.2%}'.format(tpr))
    print(label, ': FPR = FP / (FP + TN)：', '{:.2%}'.format(fpr))
    
countResult('train', resultTrain)
countResult('val', resultVal)
countResult('test', resultTest)

toPrintDF.to_csv('resource/result_data.csv', index=None)
toPrintDF
#生成6维数据（每组传感器有两个数据特征）
```
